```python
# Restaurant Review Analysis - Sentiment Analysis and NLP
## Natural Language Processing & Emotional Insights
# Project Overview
"""This notebook performs advanced sentiment analysis and NLP on restaurant review texts to extract emotional insights and patterns."""
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# NLP Libraries
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
# Text processing and visualization
from wordcloud import WordCloud
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Setup plotting
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)

# Download NLTK resources
try:
    nltk.data.find('vader_lexicon')
except LookupError:
    nltk.download('vader_lexicon')

try:
    nltk.data.find('stopwords')
except LookupError:
    nltk.download('stopwords')

try:
    nltk.data.find('punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

try:
    nltk.data.find('wordnet')
except LookupError:
    nltk.download('wordnet')

print("‚úÖ Libraries imported and NLTK resources downloaded successfully")
# Load data\n",
reviews = pd.read_csv("../../data/processed/cleaned/reviews_clean.csv")
restaurants = pd.read_csv("../../data/processed/cleaned/restaurants_clean.csv")

# Convert dates
reviews['review_date'] = pd.to_datetime(reviews['review_date'])
print("üìä Data Loaded Successfully:")
print(f"   Reviews: {reviews.shape}")
print(f"   Restaurants: {restaurants.shape}")
print(f"\nüìù Sample Review Texts:")
for i, text in enumerate(reviews['review_text'].head(3)):
    print(f"   {i+1}. {text[:100]}...")
## 1. Text Preprocessing"
class TextPreprocessor:
    def __init__(self):
        self.stop_words = set(stopwords.words('english'))
        self.lemmatizer = WordNetLemmatizer()
        # Custom stop words for restaurant context
        self.custom_stop_words = {'restaurant', 'food', 'place', 'time', 'service', 'order', 
                                'menu', 'table', 'server', 'waitress', 'waiter', 'meal'}
        self.stop_words.update(self.custom_stop_words)
    
    def clean_text(self, text):
        """Clean and preprocess text"""
        if pd.isna(text):
            return ""
        
        # Convert to lowercase
        text = text.lower()
        
        # Remove URLs
        text = re.sub(r'http\\S+|www\\.\\S+', '', text)
        
        # Remove punctuation and numbers
        text = re.sub(r'[^a-zA-Z\\s]', '', text)
        
        # Remove extra whitespace
        text = ' '.join(text.split())
        
        return text

    def tokenize_and_lemmatize(self, text):
        """Tokenize and lemmatize text"""
        tokens = word_tokenize(text)
     
        # Remove stopwords and short tokens
        tokens = [token for token in tokens if token not in self.stop_words and len(token) > 2]
        
        # Lemmatize
        tokens = [self.lemmatizer.lemmatize(token) for token in tokens]
  
        return tokens

# Initialize preprocessor
preprocessor = TextPreprocessor()

# Apply text cleaning and tokenization
print("üßπ Cleaning and preprocessing review texts...")
reviews['cleaned_text'] = reviews['review_text'].apply(preprocessor.clean_text)
reviews['tokens'] = reviews['cleaned_text'].apply(preprocessor.tokenize_and_lemmatize)
reviews['processed_text'] = reviews['tokens'].apply(lambda x: ' '.join(x))
# Remove empty reviews after cleaning
original_count = len(reviews)
reviews = reviews[reviews['processed_text'].str.len() > 10]
filtered_count = len(reviews)
print(f"‚úÖ Text preprocessing completed!")
print(f"   Original reviews: {original_count:,}")
print(f"   After filtering: {filtered_count:,}")
print(f"   Removed: {original_count - filtered_count:,} empty/short reviews")

print(f"\nüìù Text Preprocessing Examples:")
sample_reviews = reviews.head(3)
for idx, row in sample_reviews.iterrows():
    print(f"\nReview {idx}:")
    print(f"   Original: {row['review_text'][:100]}...")
    print(f"   Cleaned: {row['processed_text'][:100]}...")
## 2. Sentiment Analysis with VADER"
print("üòä Performing Sentiment Analysis with VADER...\n")

# Initialize VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Define sentiment analysis function
def get_vader_sentiment(text):
    """Get VADER sentiment scores"""
    scores = sia.polarity_scores(text)
    return scores

# Apply sentiment analysis
sentiment_scores = reviews['processed_text'].apply(get_vader_sentiment)
sentiment_df = pd.DataFrame(sentiment_scores.tolist())

# Add sentiment scores to reviews
reviews = pd.concat([reviews, sentiment_df], axis=1)

# Categorize sentiment based on compound score
def categorize_sentiment(compound):
    if compound >= 0.05:
        return 'Positive'
    elif compound <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

reviews['vader_sentiment'] = reviews['compound'].apply(categorize_sentiment)

# Also use TextBlob for comparison
def get_textblob_sentiment(text):
    """Get TextBlob sentiment"""
    blob = TextBlob(text)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

textblob_scores = reviews['processed_text'].apply(get_textblob_sentiment)
reviews['textblob_polarity'] = textblob_scores.apply(lambda x: x[0])
reviews['textblob_subjectivity'] = textblob_scores.apply(lambda x: x[1])

print("‚úÖ Sentiment analysis completed!")
print(f"\nüìä VADER Sentiment Distribution:")
print(reviews['vader_sentiment'].value_counts())
print(f"\nüìà Sentiment Score Statistics:")
print(f"   VADER Compound: {reviews['compound'].mean():.3f} ¬± {reviews['compound'].std():.3f}")
print(f"   TextBlob Polarity: {reviews['textblob_polarity'].mean():.3f} ¬± {reviews['textblob_polarity'].std():.3f}")
print(f"   TextBlob Subjectivity: {reviews['textblob_subjectivity'].mean():.3f} ¬± {reviews['textblob_subjectivity'].std():.3f}")

# Compare VADER with actual ratings
print(f"\nüîç Sentiment vs Rating Comparison:")
for sentiment in ['Positive', 'Neutral', 'Negative']:
    avg_rating = reviews[reviews['vader_sentiment'] == sentiment]['rating'].mean()
    print(f"   {sentiment} sentiment: {avg_rating:.2f}/5 average rating")
## 3. Sentiment Visualization & Analysis"
print("üìä Visualizing Sentiment Analysis Results...\n")

# Create comprehensive sentiment visualizations
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 1. VADER sentiment distribution
sentiment_counts = reviews['vader_sentiment'].value_counts()
axes[0,0].pie(sentiment_counts.values, labels=sentiment_counts.index, autopct='%1.1f%%', 
              colors=['lightgreen', 'lightgray', 'lightcoral'])
axes[0,0].set_title('VADER Sentiment Distribution', fontsize=14, fontweight='bold')
# 2. Compound score distribution
axes[0,1].hist(reviews['compound'], bins=50, alpha=0.7, color='skyblue', edgecolor='black')
axes[0,1].axvline(x=0.05, color='red', linestyle='--', alpha=0.7, label='Positive threshold')
axes[0,1].axvline(x=-0.05, color='red', linestyle='--', alpha=0.7, label='Negative threshold')
axes[0,1].set_title('VADER Compound Score Distribution', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Compound Score')
axes[0,1].set_ylabel('Frequency')
axes[0,1].legend()

# 3. Sentiment vs Rating
sentiment_rating = reviews.groupby('vader_sentiment')['rating'].mean()
axes[0,2].bar(sentiment_rating.index, sentiment_rating.values, 
              color=['lightcoral', 'lightgray', 'lightgreen'], alpha=0.7)
axes[0,2].set_title('Average Rating by Sentiment', fontsize=14, fontweight='bold')
axes[0,2].set_ylabel('Average Rating')
for i, v in enumerate(sentiment_rating.values):
    axes[0,2].text(i, v + 0.05, f'{v:.2f}', ha='center', va='bottom', fontweight='bold')

# 4. VADER vs TextBlob comparison
axes[1,0].scatter(reviews['compound'], reviews['textblob_polarity'], alpha=0.3, color='purple')
axes[1,0].set_xlabel('VADER Compound Score')
axes[1,0].set_ylabel('TextBlob Polarity')
axes[1,0].set_title('VADER vs TextBlob Sentiment Comparison', fontsize=14, fontweight='bold')
axes[1,0].axhline(y=0, color='red', linestyle='-', alpha=0.3)
axes[1,0].axvline(x=0, color='red', linestyle='-', alpha=0.3)

# 5. Sentiment by review length
reviews['review_length_chars'] = reviews['review_text'].str.len()
sentiment_length = reviews.groupby('vader_sentiment')['review_length_chars'].mean()
axes[1,1].bar(sentiment_length.index, sentiment_length.values,
              color=['lightcoral', 'lightgray', 'lightgreen'], alpha=0.7)
axes[1,1].set_title('Average Review Length by Sentiment', fontsize=14, fontweight='bold')
axes[1,1].set_ylabel('Average Characters per Review')
for i, v in enumerate(sentiment_length.values):
    axes[1,1].text(i, v + 5, f'{v:.0f}', ha='center', va='bottom', fontweight='bold')
# 6. Subjectivity distribution
axes[1,2].hist(reviews['textblob_subjectivity'], bins=30, alpha=0.7, color='orange', edgecolor='black')
axes[1,2].set_title('TextBlob Subjectivity Distribution', fontsize=14, fontweight='bold')
axes[1,2].set_xlabel('Subjectivity Score')
axes[1,2].set_ylabel('Frequency')
plt.tight_layout()
plt.show()

# Detailed sentiment analysis
print("\nüîç DETAILED SENTIMENT ANALYSIS:")
print(f"‚Ä¢ Sentiment-Rating Alignment: {((reviews['vader_sentiment'] == 'Positive') & (reviews['rating'] >= 4)).sum() / len(reviews) * 100:.1f}% of positive sentiment aligns with 4-5 star ratings")
print(f"‚Ä¢ Sentiment-Rating Mismatch: {((reviews['vader_sentiment'] == 'Negative') & (reviews['rating'] >= 4)).sum() / len(reviews) * 100:.1f}% of negative sentiment has 4-5 star ratings")
print(f"‚Ä¢ Average subjectivity: {reviews['textblob_subjectivity'].mean():.3f} (0=factual, 1=opinionated)")
print(f"‚Ä¢ Positive reviews are {reviews[reviews['vader_sentiment'] == 'Positive']['textblob_subjectivity'].mean() - reviews['textblob_subjectivity'].mean():.3f} more subjective than average")
print(f"\nüìà Sentiment by Component Ratings:")
for component in ['food_rating', 'service_rating', 'ambiance_rating']:
    sentiment_component = reviews.groupby('vader_sentiment')[component].mean()
    print(f"   {component}:")
    for sentiment, score in sentiment_component.items():
        print(f"     {sentiment}: {score:.2f}/5")
## 4. Word Cloud Analysis"
print("‚òÅÔ∏è Generating Word Clouds by Sentiment...\n")

# Combine tokens for each sentiment
positive_tokens = []
for tokens in reviews[reviews['vader_sentiment'] == 'Positive']['tokens']:
    if isinstance(tokens, list):
        positive_tokens.extend(tokens)

negative_tokens = []
for tokens in reviews[reviews['vader_sentiment'] == 'Negative']['tokens']:
    if isinstance(tokens, list):
        negative_tokens.extend(tokens)

neutral_tokens = []
for tokens in reviews[reviews['vader_sentiment'] == 'Neutral']['tokens']:
    if isinstance(tokens, list):
        neutral_tokens.extend(tokens)

# Check if wordcloud is available
try:
    _ = WordCloud
    WORDCLOUD_AVAILABLE = True
except NameError:
    WORDCLOUD_AVAILABLE = False

# Create word clouds
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

if WORDCLOUD_AVAILABLE:
    # All reviews word cloud
    all_text = ' '.join(reviews['processed_text'])
    wordcloud_all = WordCloud(width=800, height=400, background_color='white', 
                             max_words=100, colormap='viridis').generate(all_text)
    axes[0,0].imshow(wordcloud_all, interpolation='bilinear')
    axes[0,0].set_title('All Reviews - Word Cloud', fontsize=14, fontweight='bold')
    axes[0,0].axis('off')

    # Positive reviews word cloud
    positive_text = ' '.join(reviews[reviews['vader_sentiment'] == 'Positive']['processed_text'])
    if positive_text.strip():
        wordcloud_positive = WordCloud(width=800, height=400, background_color='white',
                                      max_words=100, colormap='Greens').generate(positive_text)
        axes[0,1].imshow(wordcloud_positive, interpolation='bilinear')
        axes[0,1].set_title('Positive Reviews - Word Cloud', fontsize=14, fontweight='bold')
        axes[0,1].axis('off')
    else:
        axes[0,1].text(0.5, 0.5, "No positive reviews to plot", ha='center', va='center', fontsize=12)
        axes[0,1].axis('off')

    # Negative reviews word cloud
    negative_text = ' '.join(reviews[reviews['vader_sentiment'] == 'Negative']['processed_text'])
    if negative_text.strip():
        wordcloud_negative = WordCloud(width=800, height=400, background_color='white',
                                      max_words=100, colormap='Reds').generate(negative_text)
        axes[1,0].imshow(wordcloud_negative, interpolation='bilinear')
        axes[1,0].set_title('Negative Reviews - Word Cloud', fontsize=14, fontweight='bold')
        axes[1,0].axis('off')
    else:
        axes[1,0].text(0.5, 0.5, "No negative reviews to plot", ha='center', va='center', fontsize=12)
        axes[1,0].axis('off')

    # Neutral reviews word cloud
    neutral_text = ' '.join(reviews[reviews['vader_sentiment'] == 'Neutral']['processed_text'])
    wordcloud_neutral = WordCloud(width=800, height=400, background_color='white',
                                  max_words=100, colormap='Blues').generate(neutral_text)
    axes[1,1].imshow(wordcloud_neutral, interpolation='bilinear')
    axes[1,1].set_title('Neutral Reviews - Word Cloud', fontsize=14, fontweight='bold')
    axes[1,1].axis('off')
else:
    axes[0,0].text(0.5, 0.5, "wordcloud not available\nInstall 'wordcloud'", ha='center', va='center', fontsize=12)
    axes[0,0].axis('off')
    axes[0,1].axis('off')
    axes[1,0].axis('off')
    axes[1,1].axis('off')

plt.tight_layout()
plt.show()

# Analyze most frequent words by sentiment
print("\nüìä Top Words by Sentiment:")
def get_top_words(tokens, n=10):
    """Get top n words from token list"""
    word_freq = Counter(tokens)
    return word_freq.most_common(n)

print("\n‚úÖ Positive Reviews:")
for word, count in get_top_words(positive_tokens):
    print(f"   {word}: {count}")

print("\n‚ùå Negative Reviews:")
for word, count in get_top_words(negative_tokens):
    print(f"   {word}: {count}")

print("\n‚öñÔ∏è  Neutral Reviews:")
for word, count in get_top_words(neutral_tokens):
    print(f"   {word}: {count}")

# Sentiment-specific insights
positive_words_set = set([word for word, count in get_top_words(positive_tokens, 20)])
negative_words_set = set([word for word, count in get_top_words(negative_tokens, 20)])
unique_positive_words = positive_words_set - negative_words_set
unique_negative_words = negative_words_set - positive_words_set

print(f"\nüîç Unique Positive Words: {', '.join(list(unique_positive_words)[:10])}")
print(f"üîç Unique Negative Words: {', '.join(list(unique_negative_words)[:10])}")
## 5. Topic Modeling with LDA"
print("üîç Performing Topic Modeling with LDA...")

# Prepare data for topic modeling
documents = reviews['processed_text'].tolist()

# Create document-term matrix
vectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=1000, stop_words='english')
doc_term_matrix = vectorizer.fit_transform(documents)

# Get feature names
feature_names = vectorizer.get_feature_names_out()

# Apply LDA
num_topics = 5
lda_model = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda_output = lda_model.fit_transform(doc_term_matrix)
# Display topics
def display_topics(model, feature_names, no_top_words):
    topics = {}
    for topic_idx, topic in enumerate(model.components_):
        top_words = [feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]
        topics[f"Topic {topic_idx + 1}"] = top_words
    return topics

topics = display_topics(lda_model, feature_names, 10)

print("\nüìö Discovered Topics:")
for topic_name, words in topics.items():
    print(f"\n{topic_name}: {', '.join(words)}")

# Assign dominant topic to each review
reviews['dominant_topic'] = lda_output.argmax(axis=1) + 1
print("\n‚úÖ Topic modeling completed!")
print(f"\nüìä Topic Distribution by Sentiment:")
topic_sentiment = pd.crosstab(reviews['dominant_topic'], reviews['vader_sentiment'])
print(topic_sentiment)
# Topic visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Topic distribution
topic_counts = reviews['dominant_topic'].value_counts().sort_index()
axes[0].bar(topic_counts.index, topic_counts.values, color='lightblue', alpha=0.7)
axes[0].set_title('Topic Distribution Across Reviews', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Topic Number')
axes[0].set_ylabel('Number of Reviews')
for i, v in enumerate(topic_counts.values):
    axes[0].text(i + 1, v + 10, str(v), ha='center', va='bottom')

# Sentiment by topic
topic_sentiment_percent = topic_sentiment.div(topic_sentiment.sum(axis=1), axis=0)
topic_sentiment_percent.plot(kind='bar', stacked=True, ax=axes[1], 
                           color=['lightcoral', 'lightgray', 'lightgreen'])
axes[1].set_title('Sentiment Distribution by Topic', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Topic Number')
axes[1].set_ylabel('Proportion')
axes[1].legend(title='Sentiment')
plt.tight_layout()
plt.show()

# Topic insights
print("\nüí° TOPIC ANALYSIS INSIGHTS:")
for topic_num in range(1, num_topics + 1):
    topic_reviews = reviews[reviews['dominant_topic'] == topic_num]
    avg_rating = topic_reviews['rating'].mean()
    sentiment_dist = topic_reviews['vader_sentiment'].value_counts(normalize=True)
    
    print(f"\nüìñ Topic {topic_num}:")
    print(f"   Key words: {', '.join(topics[f'Topic {topic_num}'][:5])}")
    print(f"   Average rating: {avg_rating:.2f}/5")
    print(f"   Sentiment: {sentiment_dist.get('Positive', 0)*100:.1f}% positive")
    print(f"   Sample review: {topic_reviews['review_text'].iloc[0][:100]}...")
## 6. Aspect-Based Sentiment Analysis"
print("üéØ Performing Aspect-Based Sentiment Analysis...\n")

# Define aspect categories
aspect_keywords = {
    'food_quality': ['tasty', 'delicious', 'flavor', 'fresh', 'cooked', 'quality', 'taste'],
    'service': ['service', 'friendly', 'attentive', 'efficient', 'knowledgeable', 'helpful'],
    'ambiance': ['atmosphere', 'ambiance', 'decor', 'music', 'lighting', 'clean', 'comfortable'],
    'price_value': ['price', 'value', 'expensive', 'affordable', 'worth', 'cost', 'money'],
    'wait_time': ['wait', 'time', 'reservation', 'busy', 'crowded', 'quick', 'slow']
}

# Function to detect aspects in text
def detect_aspects(text, aspects_dict):
    detected_aspects = []
    text_lower = text.lower()
    
    for aspect, keywords in aspects_dict.items():
        for keyword in keywords:
            if keyword in text_lower:
                detected_aspects.append(aspect)
                break  # Only count each aspect once per review

    return list(set(detected_aspects))

# Apply aspect detection
reviews['detected_aspects'] = reviews['review_text'].apply(
    lambda x: detect_aspects(x, aspect_keywords)
)

# Count aspect mentions
aspect_mentions = {}
for aspect in aspect_keywords.keys():
    aspect_mentions[aspect] = reviews['detected_aspects'].apply(
        lambda x: aspect in x
    ).sum()
print("\nüìä Aspect Mention Frequency:")
for aspect, count in aspect_mentions.items():
    percentage = (count / len(reviews)) * 100
    print(f"   {aspect.replace('_', ' ').title()}: {count} mentions ({percentage:.1f}% of reviews)")
print("\nüòä Aspect-Based Sentiment Analysis:")
aspect_sentiment = {}

for aspect in aspect_keywords.keys():
    aspect_reviews = reviews[reviews['detected_aspects'].apply(lambda x: aspect in x)]
    if len(aspect_reviews) > 0:
        avg_sentiment = aspect_reviews['compound'].mean()
        avg_rating = aspect_reviews['rating'].mean()
        aspect_sentiment[aspect] = {
            'avg_sentiment': avg_sentiment,
            'avg_rating': avg_rating,
            'count': len(aspect_reviews)
        }
        sentiment_label = 'Positive' if avg_sentiment > 0.05 else 'Negative' if avg_sentiment < -0.05 else 'Neutral'
        print(f"\n   {aspect.replace('_', ' ').title()}:")
        print(f"     Mentions: {len(aspect_reviews)} reviews")
        print(f"     Average sentiment: {avg_sentiment:.3f} ({sentiment_label})")
        print(f"     Average rating: {avg_rating:.2f}/5")

# Aspect sentiment visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Aspect mention frequency
aspect_names = [key.replace('_', '\n').title() for key in aspect_mentions.keys()]
mention_counts = list(aspect_mentions.values())
axes[0].bar(aspect_names, mention_counts, color='lightblue', alpha=0.7)
axes[0].set_title('Aspect Mention Frequency', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Number of Mentions')
axes[0].tick_params(axis='x', rotation=45)
for i, v in enumerate(mention_counts):
    axes[0].text(i, v + 10, str(v), ha='center', va='bottom')

# Aspect sentiment
aspect_sentiment_values = [
    aspect_sentiment[aspect]['avg_sentiment'] if aspect in aspect_sentiment else 0
    for aspect in aspect_keywords.keys()
]
colors = ['lightcoral' if x < -0.05 else 'lightgreen' if x > 0.05 else 'lightgray' for x in aspect_sentiment_values]
axes[1].bar(aspect_names, aspect_sentiment_values, color=colors, alpha=0.7)
axes[1].axhline(y=0.05, color='green', linestyle='--', alpha=0.5, label='Positive threshold')
axes[1].axhline(y=-0.05, color='red', linestyle='--', alpha=0.5, label='Negative threshold')
axes[1].set_title('Average Sentiment by Aspect', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Average Sentiment Score')
axes[1].tick_params(axis='x', rotation=45)
axes[1].legend()

plt.tight_layout()
plt.show()

# Aspect-based insights
print("\nüí° ASPECT-BASED INSIGHTS:")
if aspect_sentiment:
    most_positive_aspect = max(aspect_sentiment.items(), key=lambda x: x[1]['avg_sentiment'])
    most_negative_aspect = min(aspect_sentiment.items(), key=lambda x: x[1]['avg_sentiment'])
    most_discussed_aspect = max(aspect_mentions.items(), key=lambda x: x[1])

    print(f"‚Ä¢ Most positive aspect: {most_positive_aspect[0].replace('_', ' ')} (sentiment: {most_positive_aspect[1]['avg_sentiment']:.3f})")
    print(f"‚Ä¢ Most negative aspect: {most_negative_aspect[0].replace('_', ' ')} (sentiment: {most_negative_aspect[1]['avg_sentiment']:.3f})")
    print(f"‚Ä¢ Most discussed aspect: {most_discussed_aspect[0].replace('_', ' ')} ({most_discussed_aspect[1]} mentions)")
    print(f"‚Ä¢ Customers care most about: {', '.join([k for k, v in sorted(aspect_mentions.items(), key=lambda x: x[1], reverse=True)[:3]])}")
else:
    print("No aspects detected in reviews.")
## 7. Advanced NLP Insights & Business Applications"
print("üöÄ Advanced NLP Insights & Business Applications...\n")

# Merge with restaurant data for deeper insights
reviews_with_restaurant = reviews.merge(
    restaurants[['restaurant_id', 'cuisine_type', 'price_range', 'location']], 
    on='restaurant_id', 
    how='left'
)

# 1. Sentiment by cuisine type
cuisine_sentiment = reviews_with_restaurant.groupby('cuisine_type').agg({
    'compound': 'mean',
    'rating': 'mean',
    'review_text': 'count'
    }).round(3)
cuisine_sentiment = cuisine_sentiment.rename(columns={'review_text': 'review_count'})
cuisine_sentiment = cuisine_sentiment.sort_values('compound', ascending=False)

print("\nüçï Sentiment by Cuisine Type (Top 10):")
print(cuisine_sentiment.head(10))

# 2. Sentiment by price range
price_sentiment = reviews_with_restaurant.groupby('price_range').agg({
    'compound': 'mean',
    'rating': 'mean',
    'review_text': 'count'
    }).round(3)
price_sentiment = price_sentiment.rename(columns={'review_text': 'review_count'})

print("\nüí∞ Sentiment by Price Range:")
print(price_sentiment)

# 3. Emotional analysis - detect emotional words
emotional_words = {
    'positive_emotion': ['love', 'amazing', 'excellent', 'fantastic', 'wonderful', 'perfect', 'outstanding'],
    'negative_emotion': ['hate', 'terrible', 'awful', 'horrible', 'disappointing', 'disgusting', 'never'],
    'surprise': ['surprised', 'unexpected', 'shocked', 'astonishing', 'remarkable'],
    'anger': ['angry', 'frustrated', 'mad', 'outraged', 'infuriated']
}

def detect_emotions(text, emotion_dict):
    emotions = {}
    text_lower = text.lower()
    
    for emotion, words in emotion_dict.items():
        count = sum(1 for word in words if word in text_lower)
        emotions[emotion] = count
    
    return emotions

# Apply emotion detection
emotion_results = reviews['review_text'].apply(lambda x: detect_emotions(x, emotional_words))
emotion_df = pd.DataFrame(emotion_results.tolist())
reviews_with_emotions = pd.concat([reviews, emotion_df], axis=1)

# Emotion analysis
print("\nüòä Emotional Word Frequency:")
for emotion in emotional_words.keys():
    total_mentions = reviews_with_emotions[emotion].sum()
    percentage = (total_mentions / len(reviews)) * 100
    print(f"   {emotion.replace('_', ' ').title()}: {total_mentions} mentions ({percentage:.1f}% of reviews)")

# Create comprehensive insights visualization
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Cuisine sentiment
top_cuisines = cuisine_sentiment.head(8)
axes[0,0].barh(range(len(top_cuisines)), top_cuisines['compound'], 
    color=['lightgreen' if x > 0 else 'lightcoral' for x in top_cuisines['compound']])
axes[0,0].set_yticks(range(len(top_cuisines)))
axes[0,0].set_yticklabels(top_cuisines.index)
axes[0,0].set_title('Top Cuisines by Sentiment Score', fontsize=12, fontweight='bold')
axes[0,0].set_xlabel('Average Sentiment Score')

# 2. Price range sentiment
price_colors = ['gold', 'orange', 'red', 'darkred']
axes[0,1].bar(price_sentiment.index, price_sentiment['compound'], 
              color=price_colors, alpha=0.7)
axes[0,1].set_title('Sentiment by Price Range', fontsize=12, fontweight='bold')
axes[0,1].set_ylabel('Average Sentiment Score')
for i, v in enumerate(price_sentiment['compound']):
    axes[0,1].text(i, v + 0.01, f'{v:.3f}', ha='center', va='bottom', fontweight='bold')

# 3. Emotion distribution
emotion_totals = reviews_with_emotions[list(emotional_words.keys())].sum()
axes[1,0].bar(emotion_totals.index, emotion_totals.values, 
              color=['lightgreen', 'lightcoral', 'lightyellow', 'pink'])
axes[1,0].set_title('Emotional Word Frequency', fontsize=12, fontweight='bold')
axes[1,0].set_ylabel('Total Mentions')
axes[1,0].tick_params(axis='x', rotation=45)
for i, v in enumerate(emotion_totals.values):
    axes[1,0].text(i, v + 5, str(v), ha='center', va='bottom')

# 4. Topic-sentiment correlation
topic_sentiment_corr = reviews.groupby('dominant_topic')['compound'].mean()
axes[1,1].bar(topic_sentiment_corr.index, topic_sentiment_corr.values,
              color=['lightblue' if x > 0 else 'lightcoral' for x in topic_sentiment_corr.values])
axes[1,1].set_title('Average Sentiment by Topic', fontsize=12, fontweight='bold')
axes[1,1].set_xlabel('Topic Number')
axes[1,1].set_ylabel('Average Sentiment Score')
for i, v in enumerate(topic_sentiment_corr.values):
    axes[1,1].text(i + 1, v + 0.01, f'{v:.3f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Final comprehensive insights\n",
print("\nüéØ COMPREHENSIVE SENTIMENT INSIGHTS:")
print(f"\nüìà Overall Sentiment Health:")
positive_pct = (reviews['vader_sentiment'] == 'Positive').sum() / len(reviews) * 100
negative_pct = (reviews['vader_sentiment'] == 'Negative').sum() / len(reviews) * 100
print(f"‚Ä¢ Positive sentiment: {positive_pct:.1f}% of reviews")
print(f"‚Ä¢ Negative sentiment: {negative_pct:.1f}% of reviews")
print(f"‚Ä¢ Net sentiment score: {positive_pct - negative_pct:.1f}%")

print(f"\nüçΩÔ∏è  Cuisine Performance:")
best_cuisine = cuisine_sentiment.iloc[0]
worst_cuisine = cuisine_sentiment.iloc[-1]
print(f"‚Ä¢ Best performing cuisine: {cuisine_sentiment.index[0]} (sentiment: {best_cuisine['compound']:.3f})")
print(f"‚Ä¢ Needs improvement: {cuisine_sentiment.index[-1]} (sentiment: {worst_cuisine['compound']:.3f})")
print(f"\nüí∞ Price Value Perception:")
best_value = price_sentiment.loc[price_sentiment['compound'].idxmax()]
print(f"‚Ä¢ Best perceived value: {price_sentiment['compound'].idxmax()} (sentiment: {best_value['compound']:.3f})")

print(f"\nüòä Emotional Engagement:")
total_emotional_mentions = emotion_totals.sum()
print(f"‚Ä¢ {total_emotional_mentions} emotional word mentions across all reviews")
print(f"‚Ä¢ Positive emotions outnumber negative by {emotion_totals['positive_emotion'] / emotion_totals['negative_emotion']:.1f}x")

print(f"\nüìö Topic Insights:")
most_positive_topic = topic_sentiment_corr.idxmax()
most_negative_topic = topic_sentiment_corr.idxmin()
print(f"‚Ä¢ Most positive topic: {most_positive_topic} (sentiment: {topic_sentiment_corr.max():.3f})")
print(f"‚Ä¢ Most concerning topic: {most_negative_topic} (sentiment: {topic_sentiment_corr.min():.3f})")
print("\n")
# Save sentiment analysis results
sentiment_results = reviews[['review_id', 'restaurant_id', 'rating', 'review_text', 
                           'compound', 'vader_sentiment', 'textblob_polarity', 
                           'textblob_subjectivity', 'dominant_topic']]

sentiment_results.to_csv(
    "../../data/processed/sentiment_analysis_results.csv", index=False
)

# Save key insights
key_sentiment_insights = {
    'overall_sentiment_score': reviews['compound'].mean(),
    'positive_review_pct': positive_pct,
    'negative_review_pct': negative_pct,
    'best_cuisine_sentiment': cuisine_sentiment.index[0],
    'best_cuisine_score': best_cuisine['compound'],
    'most_positive_aspect': most_positive_aspect[0],
    'most_negative_aspect': most_negative_aspect[0],
    'emotional_engagement_ratio': emotion_totals['positive_emotion'] / emotion_totals['negative_emotion'],
}

sentiment_insights_df = pd.DataFrame([key_sentiment_insights])
sentiment_insights_df.to_csv(
    "../../data/processed/key_sentiment_insights.csv", index=False
)

print(f"\nüíæ Sentiment analysis results saved!")
print(f"   Detailed results: ../../data/processed/sentiment_analysis_results.csv")
print(f"   Key insights: ../../data/processed/key_sentiment_insights.csv")

print("\nüéâ SENTIMENT ANALYSIS COMPLETE!")
print("\nNext: Use SQL queries for business intelligence and build interactive dashboards")
```
